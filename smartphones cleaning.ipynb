```python
def core_extractor(text):
    if pd.isna(text):
        return np.nan
    elif 'Octa Core' in text:
        return 8
    elif 'Hexa Core' in text:
        return 6
    elif 'Quad Core' in text:
        return 4
    elif 'Nine Core' in text:
        return 9
    elif 'Dual Core' in text:
        return 2
```
